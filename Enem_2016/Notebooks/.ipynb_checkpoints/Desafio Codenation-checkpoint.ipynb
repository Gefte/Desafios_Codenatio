```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
X = bos.drop('NU_NOTA_MT', axis=1)
Y = bos.NU_NOTA_MT
lm = LinearRegression()
```

```python
scores = cross_val_score(lm, X, Y, cv=10, scoring='neg_mean_squared_error')
```
